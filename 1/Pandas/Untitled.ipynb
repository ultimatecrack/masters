```python
# Defining the three dataframes indicating the gold, silver, and bronze medal counts
# of different countries
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )
```

```python
final_data = gold.merge(silver, on='Country', how='outer')
final_data = final_data.merge(bronze, on='Country', how='outer')
```

```python
final_data.rename(columns={'Medals_x':'gold', 'Medals_y':'silver', 'Medals':'bronze'}, inplace=True)
```

```python
final_data.fillna(0, inplace=True)
```

```python
final_data['Medals'] = final_data['gold'] + final_data['silver'] + final_data['bronze']
```

```python
final_data.sort_values(['Medals'], ascending=False)
```

```python
final_data['Medals'] = final_data['Medals'].astype('int')
```
